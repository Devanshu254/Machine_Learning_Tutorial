```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/content/insurance_data.csv')
```

```python
df.head()
```

```python
plt.scatter(df['age'], df['bought_insurance'], marker = '+', color = 'red')
```

```python
# we will use train test split in order to split our dataset.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df['bought_insurance'], train_size = 0.9)
```

```python
X_test
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

The model is 100% accurate because our data size if very very small.

```python
model.predict_proba(X_test)
# Simply means for first person 92% possibility that he will not buy the insurence and 07% possibility that he will
# buy the insurence, that is why the model predicted 0 means he will not buy.
```

```python
model.predict([[35]])
```

```python
model.predict_proba([[35]])
```
