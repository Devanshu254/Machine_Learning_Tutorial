We are going to solve a classification problem using decision tree today. If we have a simple dataset then we can use logistic regression but if our dataset is little complex then we might have to split our dataset again and again to come up with decision boundary and this is what decision tree algorithm will do for you.

We will use a dataset where we will try to predict if a person salary is more than 100 thousand dollar based on company title, job, degree that he has.

How we will exactly select the ordering of the features ?
Answer -> Based on entropy (measure of randomness) and whereever we will get high information gain we would use that approach at every split.
Gini impurity -> This is just impurity in dataset.

```python
import pandas as pd
df = pd.read_csv('/content/salaries.csv')
df.head()
```

```python
# I want to divide the target variable and independent variable.
inputs = df.drop(['salary_more_then_100k'], axis = 'columns')
target = df['salary_more_then_100k']
```

```python
inputs.head()
```

```python
target.head()
```

```python
# Machine learning algorithms can only understand numbers not categorical data.
# So we need to convert these particular columns into numbers -> Label Encoder.

from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])
inputs.head()
```

```python
inputs_n = inputs.drop(['company', 'job', 'degree'], axis = 'columns')
```

```python
inputs_n.head()
```

```python
X = inputs_n[['company_n', 'job_n', 'degree_n']]
y = target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```
